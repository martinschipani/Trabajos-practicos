#**Voting**

```python
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb
from sklearn.ensemble import VotingClassifier
```

```python
if not exists('voting.joblib'):
  tree_clf = tree.DecisionTreeClassifier(min_samples_split=34, min_samples_leaf= 7, max_depth= 25, criterion= 'gini', ccp_alpha= 0.0001)
  xgboost_clf = xgb.XGBClassifier(max_depth=15, learning_rate=0.1, n_estimators=100, colsample_bytree=0.8, gamma=0.1, objective='binary:logistic', random_state=42)
  rf_clf = RandomForestClassifier(criterion= 'entropy', min_samples_leaf= 1, min_samples_split= 5, n_estimators= 90)


  vot_clf = VotingClassifier(estimators=[('tree', tree_clf), ('xgb', xgboost_clf), ('rf', rf_clf)], voting='hard')

  vot_clf.fit(x_train, y_train)

  pred_vot = vot_clf.predict(x_test)
  accuracy_score(y_test, pred_vot)
  joblib.dump(vot_clf, 'voting.joblib')
```

```python
vot_clf = joblib.load('voting.joblib')
```

```python
y_pred_vot_clf = vot_clf.predict(x_test)
```

```python
tabla=confusion_matrix(y_test, y_pred_vot_clf)
sns.heatmap(tabla, cmap='Blues', annot=True, fmt='g')
plt.xlabel('Predict')
plt.ylabel('True')
```

```python
accuracyVClf= accuracy_score(y_test, y_pred_vot_clf)
recallVClf = recall_score(y_test, y_pred_vot_clf)
f1VClf = f1_score(y_test, y_pred_vot_clf)
precisionVClf = precision_score(y_test, y_pred_vot_clf)

print('Accuracy: ' + str(accuracyVClf))
print('Recall: ' + str(recallVClf))
print('f1: ' + str(f1VClf))
print('precision: ' + str(precisionVClf))
```
